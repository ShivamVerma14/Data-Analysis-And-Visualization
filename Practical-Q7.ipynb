## Q7. Consider a data frame containing data about students i.e. name, gender and passing division:

```python
import pandas as pd
import numpy as np
```

```python
frame = pd.read_excel('students.xlsx')
```

```python
frame.head()
```

### a. Perform one hot encoding of the last two columns of categorical data using the get_dummies() function.

```python
pd.get_dummies(frame, columns=['Gender', 'Pass_Division'])
```

### b. Sort this data frame on the “Birth Month” column (i.e. January to December). Hint: Convert Month to Categorical.

```python
frame['Birth_Month'] = pd.Categorical(frame['Birth_Month'], 
                                      categories=['December', 'November', 'October', 'September', 'August', 'July', 'June', 'May', 'April', 'March', 'February', 'January'],
                                      ordered=True)
frame = frame.sort_values('Birth_Month', ascending=False)
frame
```
